```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
print(df.head())
print(df.tail())
```

```python
print(df.info())
```

```python
print(df['petal_length'].mean())
print(df['petal_width'].mean())
print(df['sepal_length'].mean())
print(df['sepal_width'].mean())
```

```python
print(df['petal_length'].median())
print(df['petal_width'].median())
print(df['sepal_length'].median())
print(df['sepal_width'].median())
```

```python
print(df['petal_length'].std())
print(df['petal_width'].std())
print(df['sepal_length'].std())
print(df['sepal_width'].std())
```

```python
print(df['petal_length'].describe(),end='\n\n')
print(df['petal_width'].describe(),end='\n\n')
print(df['sepal_length'].describe(),end='\n\n')
print(df['sepal_width'].describe(),end='\n\n')
```

```python
df.hist(figsize=(10, 8), bins=15, edgecolor='black')
plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['sepal_length'], df['sepal_width'], df['petal_length'], c=pd.factorize(df['species'])[0])
```

```python
sns.pairplot(df, hue='species')
```

```python
sns.violinplot(x='species', y='sepal_length', data=df)
```
